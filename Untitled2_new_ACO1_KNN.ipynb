```python
import numpy as np
import pandas as pd

class AntColony:
    def __init__(self, num_ants, num_features, num_iterations=50, rho=0.5, alpha=1.0, beta=2.0):
        self.num_ants = num_ants
        self.num_features = num_features
        self.num_iterations = num_iterations
        self.rho = rho  # Pheromone evaporation rate
        self.alpha = alpha  # Importance of pheromone in decision making
        self.beta = beta  # Importance of heuristic information in decision making
        self.pheromone = np.ones(num_features)  # Initialize pheromone levels
        self.heuristic = np.ones(num_features)  # Initialize heuristic information

    def select_features(self):
        selected_features = []
        for _ in range(self.num_iterations):
            ant_features = []
            for _ in range(self.num_ants):
                probabilities = (self.pheromone ** self.alpha) * (self.heuristic ** self.beta)
                probabilities /= np.sum(probabilities)
                selected_feature = np.random.choice(range(self.num_features), p=probabilities)
                ant_features.append(selected_feature)
            selected_features.append(ant_features)
        return selected_features

    def update_pheromone(self, selected_features):
        for ant_features in selected_features:
            for feature in ant_features:
                self.pheromone[feature] += 1  # Increment pheromone level for selected feature
        self.pheromone *= (1 - self.rho)  # Evaporate pheromone

    def run(self):
        selected_features = self.select_features()
        self.update_pheromone(selected_features)
        return selected_features

# Load your dataset (modify the path as needed)
X_train = pd.read_excel('labelled SWaT 19- Cleaned.xlsx').drop('Attack/or not', axis=1)  # Adjust for your actual data

num_ants = 20
num_features = X_train.shape[1]
aco = AntColony(num_ants=num_ants, num_features=num_features)
selected_features = aco.run()

# Convert the indices to column names
selected_features_df = pd.DataFrame(selected_features, columns=[f'Ant_{i}' for i in range(num_ants)])
selected_features_df = selected_features_df.applymap(lambda x: X_train.columns[x])
print(selected_features_df)
```

```python
import numpy as np
import pandas as pd

# Assuming you have already run the ACO and have the `selected_features` list
# and `X_train` is your data from which features are being selected

# Flatten the list of selected features across all ants and iterations
all_selected_indices = [feature for iteration in selected_features for feature in iteration]

# Count the frequency of each feature index being selected
feature_counts = np.bincount(all_selected_indices, minlength=num_features)

# Create a DataFrame to see the feature names and their selection counts
feature_frequency_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Frequency': feature_counts
})

# You can set a threshold to select features. Here, we use mean frequency as a threshold.
threshold = feature_frequency_df['Frequency'].mean()

# Filter features that are selected more frequently than the threshold
selected_features_by_frequency = feature_frequency_df[feature_frequency_df['Frequency'] > threshold]

print("Features selected by ACO for model training:")
print(selected_features_by_frequency)

# Now you can use these features to train your model
optimal_features = selected_features_by_frequency['Feature'].tolist()
X_train_selected = X_train[optimal_features]
```
